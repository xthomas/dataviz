```python
# Loading GeoJSON from Google Storage
# mazip = 'gs://datacamp-202518.appspot.com/data/rei/state-zip-geojson/ma_massachusetts_zip_codes_geo.min.json'

# Loading GeoJSON from local volume
geo_data_file = '/home/jupyter/data/rei/zips/ri_rhode_island_zip_codes_geo.min.json'
```

```python
usa = pd.read_csv('gs://datacamp-202518.appspot.com/data/rei/realtor/RDC_Inventory_Core_Metrics_Zip_History.csv')
ri = usa[usa["zip_name"].str.contains(',\sri', na=False)]
ri[["town","state"]] = ri.zip_name.str.split(",",expand=True)
ri.drop([ \
         'flag',\
         'zip_name',\
         'active_listing_count',\
         'active_listing_count_mm',\
         'active_listing_count_yy',\
         'median_listing_price_mm',\
         'median_listing_price_yy',\
         'median_days_on_market_mm',\
         'median_days_on_market_yy',\
         'new_listing_count',\
         'new_listing_count_mm',\
         'new_listing_count_yy',\
         'price_increased_count',\
         'price_increased_count_mm',\
         'price_increased_count_yy',\
         'price_reduced_count',\
         'price_reduced_count_mm',\
         'price_reduced_count_yy',\
         'pending_listing_count',\
         'pending_listing_count_mm',\
         'pending_listing_count_yy',\
         'pending_ratio_yy',\
         'median_square_feet',\
         'median_square_feet_mm',\
         'median_square_feet_yy',\
         'average_listing_price',\
         'average_listing_price_mm',\
         'average_listing_price_yy',\
         'total_listing_count',\
         'median_listing_price_per_square_feet_mm',\
         'median_listing_price_per_square_feet_yy',\
         'total_listing_count_mm',\
         'total_listing_count_yy',\
         'pending_ratio',\
         'pending_ratio_mm',\
         'pending_ratio_yy'], axis=1, inplace=True)
```

```python
# now let's load the Map data
import json

# Load the GeoJSON
with open(geo_data_file, 'r') as jsonFile:
    geo_data = json.load(jsonFile)
tmp = geo_data

# remove ZIP codes not in geo data
geozips = []
for i in range(len(tmp['features'])):
    if tmp['features'][i]['properties']['ZCTA5CE10'] in list(ri.postal_code.unique()):
        geozips.append(tmp['features'][i])

# creating new JSON object
new_json = dict.fromkeys(['type','features'])
new_json['type'] = 'FeatureCollection'
new_json['features'] = geozips

# save uodated JSON object
open("cleaned_geodata.json", "w").write(json.dumps(new_json, sort_keys=True, indent=4, separators=(',', ': ')))
```

```python
def map_feature_by_zipcode():
    """
    Generates a folium map of Rhode Island
    :param zipcode_data: zipcode dataset
    :param col: feature to display
    :return: m
    """

    # read updated geo data
    king_geo = r'cleaned_geodata.json'

    # Initialize Folium Map with Seattle latitude and longitude
    m = folium.Map(location=[41.83, -71.38], zoom_start=10,\
                   detect_retina=True, control_scale=False)

    # Create choropleth map
    m.choropleth(
        geo_data=king_geo,
        name='choropleth',
        data=ri,
        # col: feature of interest
        columns=['postal_code', 'median_days_on_market'],
        key_on='feature.properties.ZCTA5CE10',
        fill_color='OrRd',
        fill_opacity=0.9,
        line_opacity=0.2,
    )

    folium.LayerControl().add_to(m)

    return m
```

```python
map_feature_by_zipcode()
```
