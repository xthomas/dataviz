```python
url = (
    "https://raw.githubusercontent.com/python-visualization/folium/master/examples/data"
)
state_geo = f"{url}/us-states.json"

import folium

# initialize the map and store it in a m object
m = folium.Map(location=[40, -95], zoom_start=4)

# show the map
m
```

```python
folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=ThirdLineCalledOn,
    columns=["State", "Patients"],
    key_on="feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name="3L Patients, Live Rep Interaction, No Qinlock",
).add_to(m)

folium.LayerControl().add_to(m)

m
```
