```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def typer(x):
    if x=="Publicis" or \
       x=="WPP" or \
       x=="Interpublic" or \
       x=="Havas":    
        return 'BIGHOLDING'
    elif x=="Independent":        
        return 'INDY'
    else:        
        return 'SMALLHOLDING'
# orig = pd.read_csv('test-label.csv')
orig = pd.read_csv('2016_adage-indies_v2.1.csv')
df1 = orig.copy(deep=True)
```

<hr>
First thing is to work on a 'scratch' dataframe, and group the Holding companies into 'Big, Small, Independent'
<hr>

```python
df1.PARENT=df1.PARENT.apply(typer)
df1
```

```python
dfgrouped1=df1.groupby('PARENT').sum()
dfgrouped1
```

```python
dfgrouped2=df1.groupby[['PARENT','2015 US REVENUE']].sum()
dfgrouped2
```

```python
my_plot= dfgrouped1.unstack().plot(kind='bar',stacked=True,title="Breakdown by holding company")
my_plot.set_xlabel("Customers")
my_plot.set_ylabel("Sales")
```
