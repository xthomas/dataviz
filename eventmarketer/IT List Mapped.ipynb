```python
import plotly.tools
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
import pandas as pd
plotly.tools.set_credentials_file(username='rapp.cxs', api_key='55rlfaf8bj')
```

```python
raw='./data/itlist_mapping_03.csv'
df = pd.read_csv(raw)
df
```

```python
dfusa = df[(df["COUNTRY"]=="USA")]
staterev = dfusa.groupby(dfusa["STATE"],squeeze=True)
staterev.sum()
```

```python
staterev.count()
```

```python
staterev["ESTIMATED-REVENUE"].sum().index[0]
```

```python
staterev["ESTIMATED-REVENUE"].sum()[0]
```

```python
states = []
revenue = []
```

```python
for i in staterev["ESTIMATED-REVENUE"].sum().index.get_level_values("STATE"):
#     print (i)
    states.append(i)
```

```python
states
```

```python
for j in staterev["ESTIMATED-REVENUE"].sum():
#     print (j)
    revenue.append(j)
```

```python
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = states,
#         z = dfusa['EMPLOYEE-AVERAGE'].astype(float),
        z = revenue,
        locationmode = 'USA-states',
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            )
        ),
    ) ]
```

```python
layout = dict(
        title = 'Experential Agency Revenue by Statee<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)',
        ),
    )

fig = go.Figure(data=data, layout=layout)
```

```python
py.iplot(fig, filename='experiential-map')
```
