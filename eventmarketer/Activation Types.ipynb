```python
# Just a basic configuration
import pandas as pd
import re
```

```python
# an empty variable (array) to hold the tags
activation_types = []
```

```python
# A function to create a series for each tag in the TAGS column, cell by cell
def tagSeries(x):
#     print (x)
    tags = re.findall(r'(\w.+?)\s', x)
    for tag in tags:
        activation_types.append(tag)
```

```python
raw='./data/it-list.csv'
# raw='/Users/ganesha/Documents/src/dataviz/inc5000/csv_data/it-list.txt'
df = pd.read_csv(raw)
df
```

```python
df["TAGS"].apply(tagSeries)
```

```python
alltags = pd.DataFrame(activation_types, columns = ['types'])
typing = alltags['types'].groupby(alltags['types']).count()
typing.sort_values(ascending=False,inplace=True)
typing
```
