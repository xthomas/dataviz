```python
import pandas as pd
df = pd.read_csv('./influencer-data/Doug_Laney_tweets_v0.1.csv')
df.columns.values.tolist()
```

```python
df['text']
df['text'].describe()
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
tokenizer = RegexpTokenizer(r'\w+')

# create English stop words list
en_stop = get_stop_words('en')

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
```

```python
# list for tokenized documents in loop
texts = []
for i in df['text']:
    
    # clean and tokenize document string
    raw = i.decode('utf-8').lower()
    tokens = tokenizer.tokenize(raw)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in en_stop]
    
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    
    # add tokens to list
    texts.append(stemmed_tokens)
    
    
print texts
```

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
    
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=8, id2word = dictionary, passes=30)
```

```python
print(ldamodel.print_topics(num_topics=8, num_words=4))
```
