```python
# We'll import regular expressions to prune apart these Ad Age lists
import pandas as pd
import re
```

```python
raw='/Users/xavier/Documents/src/dataviz/AgencyRevenueModels/adage-data/adage-900_2010.csv'
adage = pd.read_csv(raw)
adage
```

```python
# A function to RETURN two variables from AGENCY-COMPANY column
def indySubstitute(x):
    parent = re.compile("\[(.*)\]")
    owned = re.compile("(.+?)\[(.*)\]")
    p = parent.search(x)
    if p:
        o = owned.search(x)
        return o.group(1),o.group(2)
    else:
        return x, "Independent"
```

```python
# Create two new Columns that take the two outputs of the function
adage['AGENCY-NAME'], adage['AGENCY-OWNER'] = zip(*adage['AGENCY-COMPANY'].map(indySubstitute))
adage
```

```python
# df.apply(lambda x: x['Dates'][11:13], axis=1).astype(float)
# wanted_data.loc[:, 'age'] = wanted_data.age.apply(lambda x: x + 1)
adage.loc[:, '2010-REVENUE'] = adage['2010-REVENUE'].apply(lambda x: x*100)
```
