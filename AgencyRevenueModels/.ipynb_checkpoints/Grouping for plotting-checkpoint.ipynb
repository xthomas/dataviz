```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
# py.tools.set_credentials_file(username='rapp.cxs', api_key='55rlfaf8bj')
%matplotlib inline
```

```python
adage = pd.read_csv('./adage-data/2016_adage-indies_v2.1.csv')
```

```python
# a typing function, to split agencies into 3 types:
def typer(x):
    if x=="Publicis" or \
       x=="WPP" or \
       x=="Interpublic" or \
       x=="Havas" or \
       x=="Omnicom":            
        return 'BIGHOLD'
    elif x=="Independent":        
        return 'INDY'
    else:        
        return 'SMALLHOLD'
```

```python
adage = adage[['PARENT','2015 US REVENUE']]
adage.PARENT = adage.PARENT.apply(typer)
adage
```

Let's understand more about that Group object

```python
breakdown = adage.groupby('PARENT')
breakdown.sum()
```

```python
labels = [i for i in breakdown.sum().index]
labels
```

```python
values = [j for j in breakdown.sum()["2015 US REVENUE"]]
values
```

```python
trace = go.Pie(labels=labels,values=values)
py.iplot([trace])
```
