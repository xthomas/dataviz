```python
# A function to PRINT the companies that are already acquired
def indyPrint(x):
    parent = re.compile("\[(.*)\]")
    p = parent.search(x)
    if p:
        print ("Owned")
    else:
        print ("Fair game ➜ " + x)
```

```python
adage["AGENCY-COMPANY"].apply(indyPrint)
```

```python
# A function to RETURN two variables from AGENCY-COMPANY column
def indyPrintTwo(x):
    parent = re.compile("\[(.*)\]")
    owned = re.compile("(.+?)\[(.*)\]")
    p = parent.search(x)
    if p:
        #colA = agency name, colB = owner
        o = owned.search(x)
        print (o.group(2))        
    else:
        #colA = agency name, colB = independent
        print ("Fair game ➜ " + x)
```

```python
adage["AGENCY-COMPANY"].apply(indyPrintTwo)
```

```python
# A function to RETURN two variables from AGENCY-COMPANY column
def indySubstitute(x):
    parent = re.compile("\[(.*)\]")
    owned = re.compile("(.+?)\[(.*)\]")
    p = parent.search(x)
    if p:
        #colA = agency name, colB = owner
        o = owned.search(x)
        #print (o.group(2))        
        return o.group(1),o.group(2)
    else:
        #colA = agency name, colB = independent
        #print ("Fair game ➜ " + x)
        return x, "Independent"
```

```python
# Create two new variables that take the two outputs of the function
#df['post_score_x2'], df['post_score_x3'] = zip(*df['postTestScore'].map(score_multipler_2x_and_3x))
adage['AGENCY-NAME'], adage['AGENCY-OWNER'] = zip(*adage['AGENCY-COMPANY'].map(indySubstitute))
adage
```
