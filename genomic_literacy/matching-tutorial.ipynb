### Not sure why is says "none" but it looks like there are no  'double entries' of OMIM

--> So now the thing to do is to overwrite (in place) anything that's not an OMIM tag with NA


```python
def only_omim_tags (x):
    # only save occurrences of OMIM tag
    res = re.findall(r"(OMIM:\d{1,6})",x)
    if res:
        return res[0]
    else:
        return "NA"
```

```python
new = df.iloc[:,7:49].applymap(only_omim_tags)
```
