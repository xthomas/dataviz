```python
import pandas as pd
import re
import numpy as np
```

```python
datadir = '/Users/xthomas/Documents/data/guide/'
acmgfile = 'ACMG_Conditions_OMIM.csv'
linksfile = 'acmg-links_03.txt'
```

```python
acmg = pd.read_csv(datadir + acmgfile)
```

```python
acmg
```

```python
def get_disease_url(x):
    res = re.findall(r"(?:<a\shref=\"(.*)\">MIM\s\d{5,6})",x)
    if res:
#         print (res)
        disease_urls.append(res)
```

```python
def get_clinvar_url(x):
    res = re.findall(r"(?:<a\shref=\"(\/.*)\">ClinVar)",x)
    if res:
#         print (res)
        clinvar_urls.append(res)
```

```python
disease_urls = []
clinvar_urls = []
myfile = open(datadir + linksfile,"r")
for line in myfile.readlines():
#     print (line)
    get_disease_url(line)
    get_clinvar_url(line)  
```
