## Now let's see if the 'groups' of PhenotypeList (disease text descriptions) match the same number of PhenotypeIDS (reference IDs)

### Starting by listing the IDs and the Decsription 'groups'

```python
def extract_phenos (x):
    res = re.findall(r"(.+?);",x)
    if res:
        return res
    else:
        return x
```

```python
def count_phenos (x):
    res = re.findall(r"(.+?);",x)
    if res:
        return len(res)
    else:
        return 1
```

```python
df
```

```python
# Create Count columns 
# one for IDs, and one for Descriptions
df["IDcount"] = df.PhenotypeIDS.apply(count_phenos)
df["DESCcount"] = df.PhenotypeList.apply(count_phenos)
```

```python
# Check that the numbers of descriptions
# always match the numbers of IDs
df.loc[(df["IDcount"] != df["DESCcount"])]
```
