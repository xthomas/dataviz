```python
# write the OMIM MATCHES TO ANEW COL
def omim_matches (x):
    # match everything before a semicolon
    res = re.findall(r"(OMIM:\d{1,6})",x)
    if res:
        return res
    else:
        return ("na")
```

```python
# Make a new column which shows the 'OMIM matches'
clinvar_stub["OMIM_IDS"] = clinvar_stub.loc[:,"PhenotypeIDS"].apply(omim_matches)
```

```python
results = clinvar_stub[['PhenotypeIDS','PhenotypeList','OMIM_IDS']]
```
