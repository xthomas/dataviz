```python
# Just a basic configuration
import pandas as pd
import re
```

```python
# raw='/Users/xavier/Documents/src/dataviz/inc5000/csv_data/2016_inc5k.txt'
raw='/Users/ganesha/Documents/src/dataviz/inc5000/csv_data/it-list.txt'
df = pd.read_csv(raw)
df
```

```python
# A function to PRINT the groups in each list
# e.g. tags, url, name
def listPrint(x):
    matches = re.findall(r'class=\"(.*)\".*href=\"(.*)\">(.+?)<',x)
    return (matches)
```

```python
df["LIST"].apply(listPrint).to_csv('/Users/ganesha/Documents/src/dataviz/inc5000/csv_data/it-list.csv', encoding='utf-8')
```
