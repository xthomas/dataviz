```python
import ijson
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 6)
pylab.rcParams.update({'font.size': 20})
```

```python
f16 = "csv_data/2016_profitable.csv"
f15 = "csv_data/2015_profitable.csv"
f14 = "csv_data/2014_profitable.csv"
f13 = "csv_data/2013_profitable.csv"
f12 = "csv_data/2012_profitable.csv"
```

```python
df16 = pd.read_csv(f16, usecols=["AGENCY"])
df16.columns = ['2016']
```

```python
df15 = pd.read_csv(f15, usecols=["AGENCY"])
df15.columns = ['2015']
```

```python
df14 = pd.read_csv(f14, usecols=["AGENCY"])
df14.columns = ['2014']
```

```python
df13 = pd.read_csv(f13, usecols=["AGENCY"])
df13.columns = ['2013']
```

```python
df12 = pd.read_csv(f12, usecols=["AGENCY"])
df12.columns = ['2012']
```

```python
fiveyears = pd.concat([df16,df15,df14,df13,df12], axis = 1)
```

```python
fiveyears
```

```python
fiveyears['y2match'] = np.where(fiveyears['2016'].isin(fiveyears['2015']), 'True', 'False')
```

```python
fiveyears['y3match'] = np.where(fiveyears['2016'].isin(fiveyears['2014']), 'True', 'False')
```

```python
fiveyears['y4match'] = np.where(fiveyears['2016'].isin(fiveyears['2013']), 'True', 'False')
```

```python
fiveyears['y5match'] = np.where(fiveyears['2016'].isin(fiveyears['2012']), 'True', 'False')
```

```python
fiveyears
```

```python
threeyears = fiveyears[(fiveyears['y2match'] == "True") & (fiveyears['y3match'] == "True")]
```

```python
fouryears = fiveyears[(fiveyears['y2match'] == "True") & (fiveyears['y3match'] == "True") & (fiveyears['y4match'] == "True")]
```
